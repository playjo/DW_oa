python 打印信息时会有限制 我们将打印的编码改成gb18030

headers表示我们的请求网页的头，对于没有headers的请求可能会被服务器判定为爬虫而拒绝提供服务

通过 from bs4 import BeautifulSoup 语句导入 BeautifulSoup

然后使用 BeautifulSoup(res.text, lxmlr’) 语句将网页源代码的字符串形式解析成了 BeautifulSoup 对象

解析成了 BeautifulSoup 对象可以较为方便的提取我们需要的信息

那么如何提取信息呢？

BeautifulSoup 为我们提供了一些方法

**find()方法**和**find_all()方法**：

- **find()** 返回符合条件的**首个**数据
- **find_all()** 返回符合条件的所有**数据

````python
import io
import sys
import requests
from bs4 import BeautifulSoup
#如果出现了乱码报错，可以修改编码形式
#sys.stdout = io.TextIOWrapper(sys.stdout.buffer,encoding='gb18030')
#
headers = {
  'user-agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_13_6) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/76.0.3809.132 Safari/537.36'
}
res = requests.get('https://book.douban.com/top250', headers=headers)
soup = BeautifulSoup(res.text[:5000], 'lxml')
print(soup.find('a'))
#<a class="nav-login" href="https://accounts.douban.com/passport/login?source=book" rel="nofollow">登录/注册</a>
print(soup.find_all('a'))
#返回一个列表 包含了所有的<a>标签
````

````python
def get_info():
    csvheader=['名称','面积','朝向','户型','位置','楼层','是否有电梯','建成时间',' 门锁','绿化']
    with open('wuhan_ziru.csv', 'a+', newline='') as csvfile:
        writer  = csv.writer(csvfile)
        writer.writerow(csvheader)
        for i in range(1,3):  #总共有50页，费时间，先爬取３页
            print('正在爬取自如第%s页'%i)
            timelist=[1,2,3]
            print('有点累了，需要休息一下啦（￢㉨￢）')
            time.sleep(random.choice(timelist))   #休息1-3秒，防止给对方服务器过大的压力！！！
            url='https://wh.ziroom.com/z/p%s/'%i
            headers = {'User-Agent': random.choice(user_agent)}   #使用随机UA
            r = requests.get(url, headers=headers)
            r.encoding = r.apparent_encoding    # 将网页内容分析出的编码赋值给根据header猜测的编码
            soup = BeautifulSoup(r.text, 'lxml')    # BS有多种解析器,lxml更加强大快速
            all_info = soup.find_all('div', class_='info-box')    # 查找定位为div且class为info-box的标签
            print('开始干活咯(๑>؂<๑）')
            for info in all_info:
                href = info.find('a')   # 再在其中查找定位为a开头的标签
                if href !=None:    # 如果标签内容不为空
                    href='https:'+href['href']   # 将该地址赋值给href
                    try:
                        print('正在爬取%s'%href)    # 如果爬取正确,将网址交给get_houes_info作进一步爬取
                        house_info=get_house_info(href)    # 获取的内容赋值给house_info
                        writer.writerow(house_info)    # 逐行写入csv文件
                    except:
                        print('出错啦，%s进不去啦( •̥́ ˍ •̀ू )'%href)
````

````python
def get_house_info(href):
    #得到房屋的信息
    time.sleep(1)
    headers = {'User-Agent': random.choice(user_agent)}
    response = requests.get(url=href, headers=headers)
    response=response.content.decode('utf-8', 'ignore')  # 先进行解码
    soup = BeautifulSoup(response, 'lxml')
    name = soup.find('h1', class_='Z_name').text   # 名称
    sinfo=soup.find('div', class_='Z_home_b clearfix').find_all('dd')
    area=sinfo[0].text   # 面积
    orien=sinfo[1].text   # 朝向
    area_type=sinfo[2].text   # 户型
    dinfo=soup.find('ul',class_='Z_home_o').find_all('li')
    location=dinfo[0].find('span',class_='va').text   # 位置
    loucen=dinfo[1].find('span',class_='va').text   # 楼层
    dianti=dinfo[2].find('span',class_='va').text    # 电梯标志
    niandai=dinfo[3].find('span',class_='va').text    # 建成时间
    mensuo=dinfo[4].find('span',class_='va').text   # 门锁
    lvhua=dinfo[5].find('span',class_='va').text   #绿化状况
    ['名称','面积','朝向','户型','位置','楼层','是否有电梯','建成时间',' 门锁','绿化']
    room_info=[name,area,orien,area_type,location,loucen,dianti,niandai,mensuo,lvhua]
    return room_info
````
